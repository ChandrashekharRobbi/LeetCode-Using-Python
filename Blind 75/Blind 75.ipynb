## Contains Duplicate

* Approach 1: Using HashSet
* Time Complexity $ O(n) $

```python
def containsDuplicate1(nums):
    mp = {}
    for i in nums:
        if i in mp:
            mp[i] += 1
        else:
            mp[i] = 1
    for i in nums:
        if mp[i] > 1:
            return True
    return False
            
```

* Approach 2: Using Nested For Loops
* Time Complexity  $ð‘‚(n ^ 2)$

```python
def containsDuplicate2(nums):
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] == nums[j]:
                return True
    return False
```

* Approach 3: Using set()
* Time Complexity  $ð‘‚(n)$

```python
class Solution:
    def containsDuplicate3(self, nums) -> bool:
        # set up an hashmap
        seen = set()
        for i in nums:
            if i in seen:
                return True
            seen.add(i)
        return False
```

* Approach 4: Using sorting
* Time Complexity  $ð‘‚(n logn)$

```python
def containsDuplicate4(self, nums) -> bool:
    # set up an sorting
    nums.sort()
    for i in range(len(nums)-1):
        if nums[i] == nums[i + 1]:
            return True
    return False
```

## Valid Anagram

* Approach 1: HashMap
* Time Complexity $ O(n) $

```python
class Solution:
    def isAnagram1(self, s: str, t: str) -> bool:
        # approach 1 using hashmap
        if len(s) != len(t):
            return False
        mp1 = {}
        mp2 = {}
        for i in s:
            if i in mp1:
                mp1[i] += 1
            else:
                mp1[i] = 1
        
        for i in t:
            if i in mp2:
                mp2[i] += 1
            else:
                mp2[i] = 1
        if mp1 == mp2:
            return True
        else:
            return False
        
```

* Approach 2: Using Hashmap Slightly improved than previous
* Time Complexity $ O(n) $

```python
class Solution:
    def isAnagram2(self, s: str, t: str) -> bool:
        # approach 1 using hashmap
        if len(s) != len(t):
            return False
        mp1 = {}
        for i in s:
            if i in mp1:
                mp1[i] += 1
            else:
                mp1[i] = 1
        for num in t:
            if num in mp1:
                mp1[num] -= 1
                if mp1[num] < 0:
                    return False
            else:
                return False
        return True
```

* Time Complexity $ O(nlogn) $
